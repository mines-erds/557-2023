## First code example - can you read this?

```python
# define the function
def C_to_F(Celsius_input):
    Fahrenheit = 9.0/5.0 * Celsius_input + 32
    text = 'Why not just use Celsius? In any case, here is your answer:'
    return(text, Fahrenheit)

# then, call that function with the input "0"
C_to_F(0)
```

This is a little bit complicated for your first look at code, but it shows you how readable `python` is. This is a function, and it does the following:

1. defines a function called `C_to_F` that converts temperature units from Celsius to Fahrenheit
1. Does the math to convert Celsius to Fahrenheit using the input value `Celsius_input`
1. Asks why you would ever want to convert from Celsius (the standard SI unit) - a geeky joke ;)
1. Returns the converted value `Fahrenheit` (and the joke text too).

The last line (`C_to_F(0)`) isn't indented, so it isnt a part of the function, but instead calls the function with Celsius input `0` and returns the Fahrenheit value `32`.

That loop instructs each cat in a group of cats to look stright, then to look left. Easy to read, right? That's the idea of pythonic loops

```python
geo_words = ["Shale", "Sandstone", "Limestone", "Granite", "Schist"]
print(geo_words[2])
```

Let's make a little loop to print out all of the variables in geo_words

```python
for i in range(len(geo_words)):
  print(geo_words[i])
```

```python
# you can also do it like this, which is much cleaner
for word in geo_words:
  print(word)
```

```python
# or like this, all on one line
for random_thing in geo_words: print(random_thing)
```

So, as you can see, it doesnt really matter what you call the iterator (i.e., the thing that you call in the loop). It is usually advantageous to use the simpler syntax that doesnt need `range` and `len`, as it's just cleaner and more 'pythonic'.

What we just did, was from 0 to the length of geo_words (5 in this case) loop through and preform the print function. If you wanted to add another word to the geo_words list, would you need to change the loop?
